# Cleaning Columns

```python
epl_df.columns = epl_df.columns.str.lower().str.replace(" ", "_")

epl_df.info()
```

# Feature engineering

```python
# Per-90 helper
def per90(stat, mins):
    return stat / (mins/90)

epl_df["goals_per90"] = per90(epl_df["goals"], epl_df["mins"])
epl_df["assists_per90"] = per90(epl_df["assists"], epl_df["mins"])
epl_df["season_xg"] = epl_df["xg"]*(epl_df["mins"]/90)
epl_df["season_xa"] = epl_df["xa"]*(epl_df["mins"]/90)
epl_df["yellow_per90"] = per90(epl_df["yellow_cards"], epl_df["mins"])
epl_df["red_per90"] = per90(epl_df["red_cards"], epl_df["mins"])
```

# Descriptive analysis

```python
# Average stats by position
stat_per_pos = epl_df.groupby("position")[["goals","assists","perc_passes_completed","matches","yellow_cards","red_cards"]].mean()

# Top 10 scorers
top_scorers = epl_df.nlargest(10, "goals")[["name","position","mins","goals","season_xg","xg",]]

# Top 10 assisters
top_assisters = epl_df.nlargest(10, "assists")[["name","position","mins","assists","season_xa","xa"]]

print(stat_per_pos)
print(top_scorers)
print(top_assisters)
```

```python
midfielders = epl_df[epl_df["position"].isin(['DF,MF','FW,MF','MF,DF','MF,FW','MF'])].copy()
midfielders.head(10)
```

# Visualisation - Best passers

```python
midfielders["pass_index"] = midfielders["passes_attempted"] * (midfielders["perc_passes_completed"]/100)
top_passers = midfielders.nlargest(10, "pass_index")[["name","position","passes_attempted","perc_passes_completed","pass_index"]]
print(top_passers)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(x="pass_index", y="name", data=top_passers)

plt.title("Top 10 Passers (Pass Volume × Accuracy)")
plt.xlabel("Pass Index")
plt.ylabel("Player")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
sns.scatterplot(
    x="passes_attempted", 
    y="perc_passes_completed", 
    data=top_passers, 
    hue="name", 
    s=150
)

# Reference lines (e.g. median values)
plt.axhline(top_passers["perc_passes_completed"].median(), color="red", linestyle="--", label="Median Accuracy")
plt.axvline(top_passers["passes_attempted"].median(), color="blue", linestyle="--", label="Median Attempts")

plt.title("Top 10 Passers: Volume vs Accuracy")
plt.xlabel("Passes Attempted")
plt.ylabel("Pass Completion (%)")
plt.legend()
plt.show()
```
